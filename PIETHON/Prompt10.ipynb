```python
import numpy as np
import matplotlib.pyplot as plt

random_numbers = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(random_numbers, bins=100, color='red')

plt.xlabel('Random Numbers (Normal Distribution)')
plt.ylabel('Frequency')
plt.title('Distribution of 1000 Random Numbers from Normal Distribution')

plt.savefig('normal_distribution_histogram.pdf')

plt.show()
```
